WORD EMBEDDINGS


1. Implement Word Embeddings on following text words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england'].



```python
import spacy
import gensim.downloader as api
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

word_vectors = api.load("glove-wiki-gigaword-100")

def explore_embeddings(word):
    similar_words = word_vectors.most_similar(word)
    print(f"Words most similar to '{word}':")
    for similar_word, similarity in similar_words:
        print(f"{similar_word} | {similarity:.2f}")

explore_embeddings("queen")
```

2. Implement Word Embeddings using PCA (Principal Component Analysis


```python
#Visualizing Word Embedding using PCA
words =['king','queen','man','woman','paris','france','london','england']
word_vecs = [word_vectors[word] for word in words]
pca = PCA(n_components=2)
result = pca.fit_transform(word_vecs)
pca = PCA(n_components=2)
word_vecs_2d = pca.fit_transform(word_vecs)
plt.figure(figsize=(8, 6))
plt.scatter(result[:,0],result[:,1],c='green')
for i, word in enumerate(words):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]))
plt.title("Word Embeddings Visualization")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal ComponentÂ 2")
plt.show()
```

```python
#Load SpaCy model for Linguistic structure and Dependency Parsing
nlp= spacy.load("en_core_web_sm")
#Example Sentence for Analysis
sentence = "The quick brown fox jumps over the lazy dog"
#parse sentence using SpaCy
doc = nlp(sentence)
#Display linguistic Structure and dependency parsing for token in docs
for token in doc:
    print(f"Word:{token.text}, POS:{token.pos_}, Dependecy:{token.dep_}, Head Word:{token.head.text}")
```
